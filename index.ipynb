## Purpose & Research Questions

This project has three equally weighted reports and two core objectives:

**Objective 1 — Technical:** Demonstrate data wrangling competency through 
the application of Python-based data cleaning, transformation, imputation, 
functional programming, unsupervised machine learning, and visualisation 
techniques across two structurally different real-world datasets.

**Objective 2 — Analytical:** Explore mental health through two 
complementary lenses — and then discover hidden patterns within them:

> *"Among people who self-report mental health symptoms — how severe is 
> depression and anxiety, and who seeks treatment?"*  
> (Report 1 — Clinical Survey Data)

> *"Among people who reach crisis point — who presents to a Western 
> Australian Emergency Department for mental health reasons, and what 
> happens to them?"*  
> (Report 2 — WA Emergency Department Data)

> *"What hidden patient profiles and axes of variation exist within the 
> WA mental health ED data — and what do they reveal that hypothesis-driven 
> analysis alone cannot?"*  
> (Report 3 — Advanced Pattern Recognition)

**Connecting insight:** These two datasets cannot be merged — they measure 
different populations using different instruments in different countries. 
Their value lies precisely in that difference: Report 1 captures the 
subclinical self-reported end of the mental health spectrum; Report 2 
captures the acute crisis end. Report 3 then uses unsupervised machine 
learning to discover structure within the crisis data that goes beyond 
predefined hypotheses. The gap between Report 1 and Report 2 represents 
the untreated middle — people who are unwell but reach neither a survey 
nor an emergency department.

## Report Summaries & Key Findings

### Report 1 — Clinical Survey Data
**Dataset:** Social Media & Mental Health Survey (Braghieri et al., 2022)  
**Records:** 579 US college students | PHQ-9 & GAD-7 clinical scales

Key findings:
- 68.4% of respondents show moderate-to-severe depression (PHQ-9 ≥ 10)
- Only 27.8% of those with clinical symptoms are receiving treatment
- The treatment gap is largest among students with the most severe symptoms
- Social media use correlates with higher depression and anxiety scores

### Report 2 — WA Emergency Department Data
**Dataset:** WA EDDC Synthetic Dataset | WA Department of Health, 2022  
**Records:** 949,612 total ED presentations — 59,287 (6.2%) mental health

Key findings:
- Aboriginal/TSI people are **5.4x overrepresented** in mental health ED 
  presentations (17.8% of MH ED vs 3.3% of WA population)
- 30.8% of mental health presentations result in hospital admission
- 23.2% involve drugs and/or alcohol — nearly double the emergency triage 
  rate compared to non-substance presentations
- Only 2.9% of mental health ED presentations are potentially avoidable — 
  challenging the assumption that MH patients misuse emergency departments
- 8.3% leave without receiving care (did not wait or left at own risk) — 
  2.5x the general ED rate

### Report 3 — Advanced Pattern Recognition
**Techniques:** Principal Component Analysis (PCA) + Non-negative Matrix 
Factorization (NMF)  
**Input:** 59,287 patients × 7 clinical features

Key findings:
- PCA reveals variance is **evenly distributed** across 7 components — 
  mental health ED presentations are genuinely multidimensional
- PC1 captures the **acuity vs avoidability axis**; PC2 captures the 
  **age vs self-harm axis**
- NMF identifies **5 distinct patient profiles** emerging from the data:
  - Profile 1 (1.4%) — High urgency, metro, non-substance
  - Profile 2 (5.1%) — Older patients, substance + admitted
  - Profile 3 (18.5%) — Young, lower acuity
  - Profile 4 (15.5%) — Substance involved, not admitted
  - Profile 5 (59.5%) — Low complexity baseline
